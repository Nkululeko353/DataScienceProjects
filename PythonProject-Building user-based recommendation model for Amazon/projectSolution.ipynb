```python
#- Recommendation Model:
    #Some of the movies hadnâ€™t been watched and therefore, are not rated by the users. 
    #Netflix would like to take this as an opportunity and build a machine learning recommendation 
    #algorithm which provides the ratings for each of the users.
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# The above libraries loaded  are important for training,testing,model evaluation and calculating mean_squared error ,r_score etc.
```

```python
amazon_data.head()
```

```python
amazon_data['Average_rating']=amazon_data.mean(numeric_only=True, axis=1) # Because we are going to use Linear Regression,
# we need to add/append a new variable at the end which is a continuos variable.This variable will be a target or dependent 
# variable.

# So I will add a new target/dependent variable Average_rating.The Average_rating is the mean/average of every row,so axis=1.
```

```python
amazon_data # Finally the Average_rating column is added to a dataset.
```

```python
amazon_data1=amazon_data.fillna(0) # But now we will need to exclude na values from the dataset for training,testing,evaluation

# purposes.
# If we dont remove the na values from the dataset we will receive error messages stating that the Na values are not recognised.

# Therefore we need to replace na values with 0.So we use the fillna(0) method.
# Finally the Na values are replaced with 0.
```

```python
amazon_data1
```

```python
X = amazon_data1.iloc[:,:206] # We need to select 206 columns which are features(Movie1 to Movie206).
# These features are also called independent variables.
```

```python
X.head()
```

```python
y = amazon_data1.iloc[:,206:] # Here we select the target/dependent variable which is Average_rating
```

```python
y.head()
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)#Divide the data into training and test data
# we use 20% for testing and 80% for training.
```

```python
#Model
regr = linear_model.LinearRegression()
```

```python
#Training
regr.fit(X_train, y_train)
```

```python
#Printing the coefficients
regr.coef_
```

```python
regr.intercept_
```

```python
y_pred = regr.predict(X_test) # Make predictions on the test data
```

```python
y_pred
```

```python
#Evaluation
r2_score(y_test, y_pred) # The r-squared has a negative correlation
# The lower the r-squared the lower the accuracy of the model
```

```python
mean_squared_error(y_test, y_pred)
```
