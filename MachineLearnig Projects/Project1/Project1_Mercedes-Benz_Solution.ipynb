```python
from sklearn.preprocessing import LabelEncoder# We first import the following library for LabelEncoder
#we Use it to Encode labels with value between 0 and n_classes-1.
```

```python
for c in mercedes_training_data .columns:
    if mercedes_training_data[c].dtype == 'object':
        lbl = LabelEncoder()
        lbl.fit(list(mercedes_training_data [c].values) + list(mercedes_testing_data[c].values))# Fit label encoder
        mercedes_training_data [c] = lbl.transform(list(mercedes_training_data [c].values))# Transform labels to normalized encoding.
        mercedes_testing_data[c] = lbl.transform(list(mercedes_testing_data[c].values))# Transform labels to normalized encoding.
```

```python
import xgboost as xgb
from sklearn.metrics import r2_score
# We import the xgboost library to Predict the test_df values
```

```python
# Dropping ID feature & creating saperate Input/Output training data.
train_y = mercedes_training_data['y'].values
y_mean = np.mean(train_y)
id_test = mercedes_testing_data['ID'].values
train = mercedes_training_data.drop(["ID"], axis=1)
test = mercedes_testing_data.drop(["ID"], axis=1)                  

def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)
```

```python
#* Perform dimensionality reduction.
# Before we perform Dimensionality Reduction we need to first import PCA,FastICA and TruncatedSVD 

from sklearn.decomposition import PCA, FastICA, TruncatedSVD
n_comp = 15
from sklearn.model_selection import cross_val_score,KFold,train_test_split
```

```python
df_train1= train.drop(["y"], axis=1) # We need to drop the y variable in a training dataset because it is 
# a target variable.
```

```python
df_train1.shape
```

```python
feature_list = df_train1.columns
```

```python
X = df_train1[feature_list] 
train_x=df_train1.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size = 0.20, random_state =0)
```

```python
# tSVD
tsvd = TruncatedSVD(n_components=n_comp, random_state=420)
tsvd_results_train = tsvd.fit_transform(train_x)
tsvd_results_test = tsvd.transform(test)
```

```python
# PCA
pca = PCA(n_components=n_comp, random_state=42)
pca2_results_train = pca.fit_transform(train_x)
pca2_results_test = pca.transform(test)
```

```python
# ICA
ica = FastICA(n_components=n_comp, random_state=42)
ica2_results_train = ica.fit_transform(train_x)
ica2_results_test = ica.transform(test)
```

```python
# Append decomposition components to datasets
for i in range(1, n_comp+1):
    train['tsvd_' + str(i)] = tsvd_results_train[:,i-1]
    test['tsvd_' + str(i)] = tsvd_results_test[:, i-1]
    
    train['pca_' + str(i)] = pca2_results_train[:,i-1]
    test['pca_' + str(i)] = pca2_results_test[:, i-1]
    
    train['ica_' + str(i)] = ica2_results_train[:,i-1]
    test['ica_' + str(i)] = ica2_results_test[:, i-1]
    
y_mean = np.mean(y_train)
```

```python
# Predict your test_df values using xgboost
```

```python
import xgboost as xgb

# Prepare dict of params for xgboost model.
xgb_params = {
    'n_trees': 500, 
    'eta': 0.005,
    'max_depth':6,
    'subsample': 0.5,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'base_score': y_mean, # base prediction = mean(target)
    'silent': 1}


# Creating DMatrices for Xgboost training
dtrain = xgb.DMatrix(df_train1, y_train)
dtest = xgb.DMatrix(test)

# xgboost, cross-validation
cv_result = xgb.cv(xgb_params,dtrain,num_boost_round=700,verbose_eval=50,show_stdv=False)

num_boost_rounds = len(cv_result)
print(num_boost_rounds)

# Train model
model = xgb.train(dict(xgb_params, silent=0), dtrain, num_boost_round=num_boost_rounds)

# Here we create a XGboost Regression Model
# This is because the Target variable is a float datatype.


# After running the code we get the Model performance
# Results

# After the 700 num_boost_round we get 
#train-rmse: 14.0183
#test-rmse: 37.3709


# The model is peforming nice and not overfitting.
```
